```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Census.csv")  # Replace "your_dataset.csv" with the path to your dataset file

# Define the pairs of variables
pairs = [("age", "income"), ("education", "income"), ("hour-per-week", "income")]

# Create three subplots for each pair of variables
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot each pair of variables with a different type of plot
for i, (x_var, y_var) in enumerate(pairs):
    if i == 0:  # Scatter plot for the first pair
        axs[i].scatter(df[x_var], df[y_var], alpha=0.5)
        axs[i].set_xlabel(x_var.capitalize())
        axs[i].set_ylabel(y_var.capitalize())
        axs[i].set_title(f"Scatter plot of {x_var.capitalize()} vs {y_var.capitalize()}")
    elif i == 1:  # Bar plot for the second pair
        education_counts = df[x_var].value_counts().sort_index()
        axs[i].bar(education_counts.index, education_counts.values)
        axs[i].set_xlabel("Education Level")
        axs[i].set_ylabel("Frequency")
        axs[i].set_title(f"Bar plot of {x_var.capitalize()} vs Frequency")
    elif i == 2:  # Box plot for the third pair
        axs[i].boxplot([df[df[y_var] == "<=50K"][x_var], df[df[y_var] == ">50K"][x_var]], labels=["<=50K", ">50K"])
        axs[i].set_xlabel("Income")
        axs[i].set_ylabel(x_var.capitalize())
        axs[i].set_title(f"Box plot of {x_var.capitalize()} across Income Levels")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Census.csv")

# Define pairs of variables for correlation analysis
pairs = [("age", "income"), ("education-num", "income"), ("hour-per-week", "income")]

# Scatter plot
plt.figure(figsize=(10, 6))
for i, pair in enumerate(pairs, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(data=df, x=pair[0], y=pair[1])
    plt.title(f"Scatter Plot: {pair[0].capitalize()} vs {pair[1].capitalize()}")
    plt.xlabel(pair[0].capitalize())
    plt.ylabel(pair[1].capitalize())
plt.tight_layout()
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
for i, pair in enumerate(pairs, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x=pair[1], y=pair[0])
    plt.title(f"Box Plot: {pair[0].capitalize()} vs {pair[1].capitalize()}")
    plt.xlabel(pair[1].capitalize())
    plt.ylabel(pair[0].capitalize())
plt.tight_layout()
plt.show()

# Bar plot
plt.figure(figsize=(10, 6))
for i, pair in enumerate(pairs, 1):
    plt.subplot(2, 2, i)
    sns.barplot(data=df, x=pair[1], y=pair[0])
    plt.title(f"Bar Plot: {pair[0].capitalize()} vs {pair[1].capitalize()}")
    plt.xlabel(pair[1].capitalize())
    plt.ylabel(pair[0].capitalize())
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Census.csv")

# Set the figure size for all plots
plt.figure(figsize=(10, 6))

# Bar Chart
plt.subplot(3, 3, 1)
sns.countplot(data=df, x='income')
plt.title('Bar Chart')

# Pie Chart
plt.subplot(3, 3, 2)
df['income'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')

# Count Plot
plt.subplot(3, 3, 3)
sns.countplot(data=df, x='income')
plt.title('Count Plot')

# Donut Chart
plt.subplot(3, 3, 4)
df['income'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.title('Donut Chart')

# Stacked Bar Chart
plt.subplot(3, 3, 5)
sns.countplot(data=df, x='income', hue='sex')
plt.title('Stacked Bar Chart')

# Tree Map
plt.subplot(3, 3, 6)
sizes = df['income'].value_counts()
labels = sizes.index
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Tree Map')

# Word Cloud (Not suitable for income categories, included for completeness)
# (Skip word cloud as it's not suitable for this type of data)

# Show all plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Census.csv")

# Set the figure size for all plots
plt.figure(figsize=(10, 8))

# Histogram
plt.subplot(3, 3, 1)
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Box Plot
plt.subplot(3, 3, 2)
sns.boxplot(data=df, y='age', color='skyblue')
plt.title('Box Plot of Age')
plt.ylabel('Age')

# Scatter Plot (using index as x-axis for simplicity)
plt.subplot(3, 3, 3)
plt.scatter(df.index, df['age'], color='skyblue', alpha=0.5)
plt.title('Scatter Plot of Age')
plt.xlabel('Index')
plt.ylabel('Age')

# Line Plot (using index as x-axis for simplicity)
plt.subplot(3, 3, 4)
plt.plot(df.index, df['age'], color='skyblue')
plt.title('Line Plot of Age')
plt.xlabel('Index')
plt.ylabel('Age')

# Bar Plot (using age groups)
plt.subplot(3, 3, 5)
age_groups = pd.cut(df['age'], bins=[0, 20, 30, 40, 50, 60, 70, 80, 90, 100])
age_groups_counts = age_groups.value_counts().sort_index()
age_groups_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Bar Plot of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Frequency')

# Heatmap (using age as both x and y for simplicity)
plt.subplot(3, 3, 6)
age_counts = df['age'].value_counts().sort_index()
sns.heatmap(age_counts.to_frame(), cmap='Blues', annot=True, fmt='d')
plt.title('Heatmap of Age Counts')
plt.xlabel('Age')
plt.ylabel('Age')

# Violin Plot
plt.subplot(3, 3, 7)
sns.violinplot(data=df, y='age', color='skyblue')
plt.title('Violin Plot of Age')
plt.ylabel('Age')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Census.csv")

# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Attributes')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.io as pio

# Load the dataset
df = pd.read_csv("Census.csv")

# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Define heatmap trace
heatmap_trace = go.Heatmap(
    z=correlation_matrix.values,
    x=correlation_matrix.columns,
    y=correlation_matrix.index,
    colorscale='RdBu',
)

# Define layout
layout = go.Layout(
    title='Dynamic Correlation Matrix',
    xaxis=dict(title='Attributes'),
    yaxis=dict(title='Attributes'),
)

# Create figure
fig = go.Figure(data=[heatmap_trace], layout=layout)

# Show plot
pio.show(fig)
```

```python
import pandas as pd
import plotly.graph_objs as go

# Load the dataset
df = pd.read_csv("Census.csv")

# Define the trace for the 3D scatter plot
scatter_trace = go.Scatter3d(
    x=df['age'],
    y=df['hour-per-week'],
    z=df['income'],  # Assuming 'income' is the third attribute
    mode='markers',
    marker=dict(
        size=5,
        color='blue',  # Set color to blue for all points
        opacity=0.8
    )
)

# Define layout
layout = go.Layout(
    title='3D Scatter Plot of Age vs Hour-per-week vs Income',
    scene=dict(
        xaxis=dict(title='Age'),
        yaxis=dict(title='Hour-per-week'),
        zaxis=dict(title='Income')
    )
)

# Create figure
fig = go.Figure(data=[scatter_trace], layout=layout)

# Show plot
fig.show()
```
